```python
%%prun
def fib(n):
    if n<2:
        return 1
    else:
        return fib(n-1)+fib(n-2)
def fib_fast(n,a=1,b=1):
    if n==1:
        return b
    else:
        return fib_fast(n-1,b,a+b)
%time fib(20)
%time fib_fast(20)
```

```python
def as_png(img,**kw):
    import io
    from matplotlib import image
    from IPython import display
    buf=io.BytesIO()
    image.imsave(buf,img,**kw)
    return buf.getvalue()
```
